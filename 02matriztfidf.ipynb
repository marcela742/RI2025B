```python

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import nltk


ruta = "/kaggle/input/01-corpus-turismo-500-txt/01_corpus_turismo_500.txt"

with open(ruta, "r", encoding="utf-8") as f:
    documentos = f.readlines()

# Crear el vectorizador 
vectorizador = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1,1)
)

# Calcular la matriz TF-IDF
tfidf_matriz = vectorizador.fit_transform(documentos)

#  Convertir a DataFrame
tfidf_df = pd.DataFrame(
    tfidf_matriz.toarray(),
    columns=vectorizador.get_feature_names_out()
)

print("✅ Matriz TF-IDF generada correctamente")
tfidf_df.head()
```
